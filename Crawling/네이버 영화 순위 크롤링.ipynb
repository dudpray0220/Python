```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
response_html = requests.get(url)
response_html
```

```python
response_html.text
```

```python
soup = BeautifulSoup(response_html.text, 'html.parser')
soup
```

```python
title_lst = soup.find_all('div',class_='tit3')
title_lst
```

```python
title_lst[0]
```

```python
tit_temp = title_lst[0].find('a')
```

```python
tit_temp
```

```python
tit_temp.text
```

```python
title = tit_temp.text
```

```python
title
```

```python
name_lst = []
for name in title_lst:
    a_tag = name.find('a')
    title = a_tag.text
    name_lst.append(title)
name_lst
```

```python
name_lst = []
for i in range(len(title_lst)) :
    movie=title_lst[i].find('a').text
    name_lst.append(movie)
name_lst 
```

```python
name_lst = []
rank = 1
rank_lst = []
for name in title_lst:
    a_tag = name.find('a')
    title = a_tag.text
    name_lst.append(title)
    rank_lst.append(rank)
    rank = rank + 1
```

```python
name_lst
```

```python
rank_lst
```

```python
movie_df = pd.DataFrame({'순위':rank_lst, '제목':name_lst})
movie_df.head()
```
