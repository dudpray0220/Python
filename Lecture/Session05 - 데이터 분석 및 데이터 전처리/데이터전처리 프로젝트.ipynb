### [데이터 통합]

```python
selloutData["YEAR"] = selloutData.YEARWEEK.str[:4]
```

### [프로젝트 실습]
refinedData 에서 →
지역, 상품, 연주차 단위로 정렬하여
sortedData 변수에 담으세요

```python
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
```

```python
sortedData = refinedSelloutData.sort_values(by=sortKey)
```

### [프로젝트 실습]
[지역, 상품, 연도별 집계]
sortedData 에서 지역, 상품 단위
판매량(QTY_NEW) 의 평균 연산 후
groupData 변수에 담으세요

이후 컬럼명을 QTY_MEAN로 변경하세요

```python
groupKey = ["REGIONID","PRODUCT","YEAR"]
```

```python
groupData = sortedData.groupby(groupKey, as_index=False).QTY.mean()
```

```python
groupData = groupData.rename(columns = {'QTY': 'QTY_MEAN'})
```

```python
groupData.head()
```

```python
groupKey = ["REGIONID"]
```

```python
groupData = sortedData.groupby(groupKey)["QTY"].agg("mean").reset_index()
```

```python
groupData.rename({"QTY":"QTY_NEW"}, inplace=True)
```

```python
groupData.sort_values("QTY", ascending=False) # 얼마만 welloutData
```

```python
groupKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
sortedData.groupby(groupKey)["QTY"].describe()
```

```python
# 이동평균 구하기 (스칼라에선 100줄, 파이썬은 1줄 ㅋㅋ ㅅㄱ)
sortedData["MAS"] = sortedData.QTY.rolling(window=5, 
                                           min_periods=1, 
                                           center=True).mean()
```

```python
sortedData
```

```python
# 열삭제
sortedData.drop(columns="QTY_NEW")
```

```python
# np.where, function
sortedData["EFFECT"] = sortedData.QTY / sortedData.MAS
```

```python
sortedData
```

```python
sortedData.YEAR.drop_duplicates()
```

```python
groupKey = ["REGIONID", "PRODUCT", "WEEK"]
```

```python
# 3개 연도의 지역, 상품, 주차별 평균 EFFECT를 뽑아내고 싶다!!
ratioData = sortedData.groupby(groupKey)[["EFFECT"]].agg("mean").reset_index()
```
