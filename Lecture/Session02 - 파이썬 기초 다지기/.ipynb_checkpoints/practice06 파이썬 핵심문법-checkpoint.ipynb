### 미션 : 소수점 둘째자리 반올림

```python
def roundFunction(inValue):    
    # Step1: 입력받은 데이터를 반올림 자리수만큼 곱해준다.
    # 함수를 개발하기위한 디버깅코드
    # inValue = 1.893430903490
    step1 = inValue * 100


    # Step2: 곱해준결과에 0.5를 더한 후 소수점이하 버림!
    step2 = int(step1 + 0.5)

    # Step3: 100으로 나누어준다.
    step3 = step2 / 100
    outValue = step3
    
    return outValue
```

### 함수 입력변수에 따라 소수점 2자리수, 3자리 등의 반올림이 가능하도록 함수를 구현하세요 (예: roundFunction(12.2225,3) -> 3번째자리 반올림)

```python
def roundFunction2 (inValue, optionValue):
    # 디버깅코드
    # inValue = 1.2225
    # optionValue = 4
    
    step1 = inValue * 10**optionValue

    # Step2: 곱해준결과에 0.5를 더한 후 소수점이하 버림!
    step2 = int(step1 + 0.5)

    # Step3: 100으로 나누어준다.
    step3 = step2 / 10**optionValue
    outValue = step3
    
    return outValue
```

```python
def myRoundFunction(inValue1, n):
    #Step1 : 입력받은 데이터를 10의 n-1승만큼 곱해준다.
    step1 = inValue1 * (10 ** (n-1))

    #Step2 : 곱해준결과에 0.5를 더한 후 소수점이하 버림!
    step2 = int(step1 + 0.5)

    #Step3 : 10의 n-1승만큼 나누어준다.
    step3 = step2 / (10 ** (n-1))
    newRound = step3

    return newRound
```

### 미션 : 소수점 둘째자리 반올림

```python
tvList2 = ['UN40EN001', 'LEDTV_UN40EN002', 'LEDTV_LEDTV_UN40EN003','ledtv_UN40EN004']

# (문제분석) 입력데이터 접두어 "LEDTV_" 이게 2번이상 또는 아예 안 나올수도 있다.
# 대문자로 표현해야함

# Step1: 모델코드를 불러온다. (접두어가 2개 1개 0개 붙을 수 있다)

for i in range(0, len(tvList2)):

    # 디버깅 코드
    loadedModel = tvList2[i]

    # Step2: 대문자로 변환시킨다.

    modelUpper = loadedModel.upper()

    # Step3: 접두어를 찾아서 하나만 붙인다. (상세로직)
    #        모델코드를 찾고 접두어를 하나만 붙인다.
    # 하나도 없을때 (clear) -> prefix + 대문자변환 기존값
    # 1개 이상일때

    preFix = "LEDTV_"
    preFixCnt = modelUpper.count(preFix)

    if preFixCnt == 0:   #로직 구현
        modelUpper = preFix + modelUpper
    elif preFixCnt >=1:
        modelUpper = preFix + modelUpper.replace(preFix,"") 

# Step4: 정상모델코드를 출력한다.

    cvtModel = modelUpper
    tvList2[i] = cvtModel
    
print(tvList2)
```

## 현재연도주차에 빼고싶은 주차를 입력하면 뺀 결과를 알려주는 함수 (조별실습 1 0329)

```python
from isoweek import Week

def preWeek(yearWeek, inputWeeks):
    yearWeek = str(yearWeek)  
    inputWeeks = int(inputWeeks)  

    # 문자열 yearWeek를 분리 (year, week)
    year = int(yearWeek[:4])   
    week = int(yearWeek[4:])   

    # inputWeeks가 week보다 클 때는 전년도로 넘어가서 전년도 week수를 더함
    # (week가 inputWeeks보다 커질때까지 계속 while문 반복)
    if week <= inputWeeks:
        while (week <= inputWeeks):
            week = week +  Week.last_week_of_year(year - 1).week
            year = year - 1;
        week = week - inputWeeks

    # inputWeeks가 week보다 작을 때는 그냥 week에다가 inputWeeks를 뺌.    
    else :
        week = week - inputWeeks;

    # week값이 10미만일 경우 앞에 문자"0"을 추가함    
    if (week < 10):
        week = str(week)
        week = "0" + week

    print(str(year) + str(week))
```
