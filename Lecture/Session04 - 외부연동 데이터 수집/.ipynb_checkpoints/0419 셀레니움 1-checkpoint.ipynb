## 실습

셀레니움을 활용해서 sparkkorea.com 퀴즈에 접근을 한다. 이후 spark 관련 퀴즈 타이틀과 링크정보를 수집하여 데이터프레임으로 변환한다.

```python
# 라이브러리 접근
from selenium import webdriver    # 셀레니움
import pandas as pd
import time
import requests 
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

# 드라이버 옵션 설정
options = webdriver.ChromeOptions()  
options.add_argument("window-size=1920x1080")

# 드라이버 위치 설정
driverLoc = "../addon/chromedriver/chromedriver.exe"
driver = webdriver.Chrome(driverLoc, options=options)
options.add_argument("window-size=1920x1080")

# 웹페이지 파싱 될 때까지 최대 3초 기다림.
driver.implicitly_wait(3)

# 브라우저 팝업
targetUrl = "https://www.sparkkorea.com/"
driver.get(targetUrl)
time.sleep(3)

# xPath로 퀴즈에 접근
try :
    quizXpath = '//*[@id="menu-item-382"]/a'
    quizBtn = driver.find_element_by_xpath(quizXpath)
    quizBtn.click()
except:
    print(e)

# 현재 있는 페이지 정보 가져오기.
finalUrl = driver.current_url
pgSource = driver.page_source
# pgSource ## 셀레니움의 역할은 여기서 끝난다. 이제 bs4의 역할

# URL 접속
resp = requests.get(url=finalUrl)
if resp.status_code == 200:
    resp.encoding = "utf-8"
    html = resp.text
    htmlObj = BeautifulSoup(html, "html.parser")
else:
    print("다시 URL 확인해봐~")

divTags = htmlObj.find(name="div", attrs={"id" : "id_spark_quiz"})
atags = divTags.findAll(name ="a")

titleList = []
hrefList = []

for i in range(0, len(atags)):
    eachTitle = atags[i].text
    eachHref = atags[i].attrs["href"]
    titleList.append(eachTitle)
    hrefList.append(eachHref)
    
quizData = pd.DataFrame(zip(titleList, hrefList), columns=["퀴즈제목", "링크주소"])
quizData.to_csv('../dataset/quizData.csv', encoding="ms949", index=False)
quizData
```
